```python
class StackADT:
  def __init__( self ):
    self.__data = list()
  
  def is_empty( self ):
    return len( self.__data ) == 0

  def get_length( self ):
    return len( self.__data )
  
  def pop( self ):
    if self.is_empty():
      print("La pila esta vacia")
    else:
      return self.__data.pop()

  def peek( self ):
    if self.is_empty():
      pass
    else:
      return self.__data[ -1 ]

  def push( self , value ):
    self.__data.append( value )

  def to_string( self ):
    for item in self.__data:
      print(f"| { item } |")

  def balanceo( self , codigo ):
    pilaControl = StackADT()
    pilaControl.push('#')
    pilaControl1 = StackADT()
    pilaControl1.push('#')

    for letra in codigo:
      if letra == '{':
        pilaControl.push("{")
      elif letra == '}':
        basura = pilaControl.pop()
      else:
        pass

    if pilaControl.peek() == '#':
      print("El codigo proporcionado está correctamente balanceado en llaves")
      
      #Buscando balanceo de parentesis
      for letra in codigo:
        if letra == '(':
          pilaControl1.push("(")
        elif letra == ')':
          basura = pilaControl1.pop()
        else:
          pass

      if pilaControl1.peek() == '#':
        print("El codigo proporcionado está correctamente balanceado en parentesis")
      else:
        print("El codigo no esta balanceado en parentesis (y) ")
      
    #Si no esta balanceado en llaves, no te dirá si esta balanceado en parentesis (se detiene)
    else:
      print("El codigo no esta balanceado en llaves {y}\nEL PROGRAMA SE DETUVO (no hará balanceo de parentesis, por no pasar el balanceo en llaves) ")
```
